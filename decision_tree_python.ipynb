```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
#from graphviz import Digraph
import pydotplus
```

```python
from sklearn import tree
```

```python
mycols=['X1','X2']
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("Toy.csv")
```

```python
cl_tree=tree.DecisionTreeClassifier(criterion='gini',random_state=1,max_depth=3)
cl_tree=cl_tree.fit(df[['X1','X2']],df['CLASS'])
pred=cl_tree.predict(df[['X1','X2']])
print(pd.crosstab(df['CLASS'],pred))
dot_data=StringIO()
export_graphviz(cl_tree,out_file=dot_data,feature_names=mycols,filled=True,rounded=True,special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
#graph.write_png('toy_tree_depth1.png')
```
